## sum test

```python
import array
import task
import numpy as np #only use for tests

def test_matrix(test = 0):
    if test == 0:
        x = array.array("d", range(100))
        y = array.array("d", range(100))
    elif test == 1:
        x = [1, 2]
        x = [float(i) for i in x]
        y = [[1, 2], [1, 2]]
        y = [[float(i) for i in j] for j in y]
    elif test == 2:
        x = [[3, 2], [1, 8]]
        x = [[float(i) for i in j] for j in x]
        y = [[3, 2], [1, 8]]
        y = [[float(i) for i in j] for j in y]
    return x, y

def ans_test(answer, tes, test = 0):
    pre = 0
    if test == 0:
        #print(answer)
        #print(tes)
        for i in range(len(x)):
            if answer[i] == tes[i]:
                pre += 1
    elif test == 1:
        for i in range(len(answer)):
            if answer[i] == tes[i]:
                pre += 1
    else:
        for i in range(len(answer)):
            for j in range(len(answer[0])):
                if answer[i][j] == tes[i][j]:
                    pre += 1
        pre /= len(x)
    pre /= len(x)
    print(pre)

x, y = test_matrix(0)
```

```python
#%timeit
#only use for tests
X = np.asarray(x)
Y = np.asarray(y)
answer = X + Y
answer
```

```python
#%%timeit
tes = task.msum(x, y)
tes
```

```python
ans_test(answer, tes, 0)
```
